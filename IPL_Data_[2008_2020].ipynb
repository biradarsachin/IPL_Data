#1.Top10 Teams with most number of wins

```python
# Calculate the value counts for the 'winner' column
value_counts = data['winner'].value_counts()

# Sort the value counts in descending order
value_counts = value_counts.sort_values(ascending=False)

# Choose the top n categories to plot
n = 10
top_categories = value_counts.head(n)

# Create a list of colors for the bars
colors = plt.cm.viridis(np.linspace(0, 1, len(top_categories)))

# Create a bar graph with different colors for each bar
plt.figure(figsize=(10, 6))
plt.bar(top_categories.index, top_categories.values, color=colors)
plt.xlabel('Winner')
plt.ylabel('Count')
plt.title('Top {} Winners'.format(n))
plt.xticks(rotation=45)
plt.show()
```

#2.Top 10 Batsmans with most number of runs from year 2008-2020

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' containing the dataset
# and 'batsman_runs' represents the runs scored by each batsman

# Group the data by batsman and calculate the total runs
runs_by_batsman = data.groupby('batsman')['batsman_runs'].sum().reset_index()

# Sort the data in descending order based on total runs
runs_by_batsman = runs_by_batsman.sort_values('batsman_runs', ascending=False)

# Choose the top 10 batsmen based on total runs
top_batsmen = runs_by_batsman.head(10)

# Define colors for the bars
colors = ['royalblue', 'orange', 'green', 'red', 'purple', 'yellow', 'brown', 'gray', 'pink', 'teal']

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_batsmen['batsman'], top_batsmen['batsman_runs'], color=colors)
plt.xlabel('Batsman')
plt.ylabel('Total Runs')
plt.title('Distribution of Runs by Batsman')
plt.xticks(rotation=45)
plt.show()

# Create a pie chart
plt.figure(figsize=(6, 6))
plt.pie(top_batsmen['batsman_runs'], labels=top_batsmen['batsman'], autopct='%1.1f%%', colors=colors)
plt.title('Distribution of Runs by Batsman')
plt.axis('equal')
plt.show()
```

#3.Top 10 Bowlers with most number of wickets from year 2008-2020

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' containing the dataset
# and 'bowler' represents the name of the bowler
# and 'is_wicket' represents whether a wicket was taken or not

# Group the data by bowler and count the number of wickets
wickets_by_bowler = data[data['is_wicket'] == 1].groupby('bowler').size().reset_index(name='wickets')

# Sort the data in descending order based on the number of wickets
wickets_by_bowler = wickets_by_bowler.sort_values('wickets', ascending=False)

# Choose the top 10 bowlers based on the number of wickets
top_bowlers = wickets_by_bowler.head(10)

# Create a bar chart
plt.figure(figsize=(10, 6))
plt.bar(top_bowlers['bowler'], top_bowlers['wickets'],color=colors)
plt.xlabel('Bowler')
plt.ylabel('Number of Wickets')
plt.title('Distribution of Wickets by Bowlers')
plt.xticks(rotation=45)
plt.show()

# Create a pie chart
plt.figure(figsize=(8, 8))
plt.pie(top_bowlers['wickets'], labels=top_bowlers['bowler'], autopct='%1.1f%%')
plt.title('Distribution of Wickets by Bowlers')
plt.axis('equal')
plt.show()
```

#7.Winner vs toss decision winners

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'data' containing the dataset
# and 'toss_winner' represents the column for the toss winner
# and 'winner' represents the column for the match winner

# Calculate the count of matches won by each team after winning the toss
toss_winner_counts = data[data['toss_winner'] == data['winner']]['toss_winner'].value_counts()

# Calculate the count of matches won by each team regardless of winning the toss
match_winner_counts = data['winner'].value_counts()

# Combine the counts into a single DataFrame
df = pd.concat([toss_winner_counts, match_winner_counts], axis=1, keys=['Matches Won after Toss Win', 'Total Matches Won'])

# Sort the DataFrame based on the total matches won in descending order
df = df.sort_values('Total Matches Won', ascending=False)

# Choose the top 10 teams based on the total matches won
top_teams = df.head(10)

# Create a stacked bar chart
top_teams.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.xlabel('Teams')
plt.ylabel('Count')
plt.title('Toss Winner vs Match Winner')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

#10) Matches Played by each team, matches won , Win-loss ratio for year 2008

1.DECCAN CHARGES 

```python
import pandas as pd

# Convert the date column to datetime format
data['date'] = pd.to_datetime(data['date'])

# Filter the data for the year 2008
data_2008 = data[data['date'].dt.year == 2008]

# Calculate the number of matches won by Deccan Chargers in 2008
dc2008won = len(data_2008[data_2008['winner'] == 'Deccan Chargers'])

# Calculate the number of matches played by Deccan Chargers in 2008
dc2008played = len(data_2008[(data_2008['team1'] == 'Deccan Chargers') | (data_2008['team2'] == 'Deccan Chargers')])

# Calculate the win-loss ratio for Deccan Chargers in 2008
dc2008wr = dc2008won / dc2008played

# Print the results
print("dc2008played:", dc2008played)
print("dc2008won:", dc2008won)
print("dc2008 win-loss ratio:", dc2008wr)
```
